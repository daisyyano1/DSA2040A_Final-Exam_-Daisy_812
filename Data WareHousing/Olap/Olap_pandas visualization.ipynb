```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

db_path = r"C:\Users\USER\OneDrive\Desktop\Data Warehousing\DSA204 final Exam\retail_dw.db"
conn = sqlite3.connect(db_path)

query = """
SELECT 
    c.ship_country AS Country,
    t.quarter AS Quarter,
    SUM(f.sales_amount) AS TotalSales
FROM fact_sales f
JOIN dim_customer c ON f.customer_id = c.customer_src
JOIN dim_time t ON f.time_id = t.rowid
GROUP BY c.ship_country, t.quarter
ORDER BY c.ship_country, t.quarter;
"""

df = pd.read_sql_query(query, conn)
conn.close()

# Pivot so countries become x-axis
pivot = df.pivot(index="Country", columns="Quarter", values="TotalSales")

pivot.plot(kind="bar", figsize=(10,6))
plt.xlabel("Country")
plt.ylabel("Total Sales")
plt.title("Total Sales by Country and Quarter")
plt.tight_layout()

plt.savefig("sales_by_country_quarter.png")
plt.show()
```
