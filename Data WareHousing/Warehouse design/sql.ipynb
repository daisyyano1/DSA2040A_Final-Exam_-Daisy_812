```python
# Task 1: Data Warehouse Design
import sqlite3

# ---------------------------------------------
# CONNECT OR CREATE SQLite DATABASE FILE
# ---------------------------------------------
conn = sqlite3.connect("retail_dw.db")
cursor = conn.cursor()

# ---------------------------------------------
# SQL STAR SCHEMA (ALL TABLES)
# ---------------------------------------------
sql_script = """
PRAGMA foreign_keys = ON;

-- Drop existing tables (optional safety)
DROP TABLE IF EXISTS SalesFact;
DROP TABLE IF EXISTS ProductDim;
DROP TABLE IF EXISTS CategoryDim;
DROP TABLE IF EXISTS CustomerDim;
DROP TABLE IF EXISTS TimeDim;

-- Time Dimension
CREATE TABLE TimeDim (
    time_id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    day INTEGER,
    month INTEGER,
    quarter INTEGER,
    year INTEGER,
    weekday INTEGER,
    is_weekend INTEGER
);

-- Customer Dimension
CREATE TABLE CustomerDim (
    customer_id INTEGER PRIMARY KEY,
    customer_name TEXT,
    gender TEXT,
    birth_date TEXT,
    age_bucket TEXT,
    city TEXT,
    state TEXT,
    country TEXT,
    email TEXT
);

-- Category Dimension
CREATE TABLE CategoryDim (
    category_id INTEGER PRIMARY KEY,
    category_name TEXT NOT NULL,
    department TEXT
);

-- Product Dimension
CREATE TABLE ProductDim (
    product_id INTEGER PRIMARY KEY,
    stock_code TEXT,
    product_name TEXT,
    category_id INTEGER,
    category_name TEXT,
    brand TEXT,
    unit_price REAL,
    FOREIGN KEY (category_id) REFERENCES CategoryDim(category_id)
);

-- Sales Fact Table
CREATE TABLE SalesFact (
    sales_id INTEGER PRIMARY KEY AUTOINCREMENT,
    invoice_no TEXT,
    time_id INTEGER NOT NULL,
    customer_id INTEGER,
    product_id INTEGER,
    quantity INTEGER,
    unit_price REAL,
    total_sales REAL,
    country TEXT,
    stock_code TEXT,
    FOREIGN KEY (time_id) REFERENCES TimeDim(time_id),
    FOREIGN KEY (customer_id) REFERENCES CustomerDim(customer_id),
    FOREIGN KEY (product_id) REFERENCES ProductDim(product_id)
);
"""

# ---------------------------------------------
# EXECUTE SQL SCRIPT
# ---------------------------------------------
cursor.executescript(sql_script)

# Save and close
conn.commit()
conn.close()

print("✔ Star schema created successfully in retail_dw.db")

```
